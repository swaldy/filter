```python
import matplotlib
import matplotlib.pyplot as plt
import os
import pandas as pd
import glob
import numpy as np
import csv
```

```python
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:purple', 'tab:orange', 'tab:olive', 'tab:brown']

results_dir = 'results_3s'
dictionary = {
    '50x10': [],
    '50x12P5': [],
    '50x15': [],
    '50x20': [],
    '50x25': [],
    '100x25': [],
    '100x25x150': []
}

results_dir2 = 'results_5s'
dictionary2 = {
    '50x10': [],
    '50x12P5': [],
    '50x15': [],
    '50x20': [],
    '50x25': [],
    '100x25': [],
    '100x25x150': []
}

legend_dict = {
    '50x10': '50x10x100',
    '50x12P5': '50x12.5x100',
    '50x15': '50x15x100',
    '50x20': '50x20x100',
    '50x25': '50x25x100',
    '100x25': '100x25x100',
    '100x25x150': '100x25x150'
}

save_dir = './custom_plots/'
# sizes = ['50x10', '50x12P5', '50x15', '50x20', '50x25', '100x25x150']
sizes = ['50x10', '50x12P5', '50x15', '50x20', '50x25', '100x25', '100x25x150']
thresholds = [0.1, 0.15, 0.2, 0.3, 0.4, 0.5]
df_temp = pd.read_csv('./'+results_dir+'/final_results.txt',names=["sensor_geom","threshold","run_iter","sig_eff","sig_eff_err","bg_rej","bg_rej_err","dat_red","dat_red_err"])
for size_iter in sizes:
    for threshold in thresholds:
        sig_eff_errStat = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["sig_eff_err"].max()
        sig_eff_errML = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["sig_eff"].std()
        sig_eff_err = np.sqrt(sig_eff_errML**2 + sig_eff_errStat**2)
        sig_eff = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["sig_eff"].mean()
        
        bg_rej_errStat = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["bg_rej_err"].max()
        bg_rej_errML = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["bg_rej"].std()
        bg_rej_err = np.sqrt(bg_rej_errML**2 + bg_rej_errStat**2)
        bg_rej = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["bg_rej"].mean()
        
        dat_red_errStat = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["dat_red_err"].max()
        dat_red_errML = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["dat_red"].std()
        dat_red_err = np.sqrt(dat_red_errML**2 + dat_red_errStat**2)
        dat_red = df_temp[(df_temp.sensor_geom==size_iter) & (df_temp.threshold==threshold)]["dat_red"].mean()
        dictionary[size_iter].append([threshold, sig_eff, sig_eff_err, bg_rej, bg_rej_err, dat_red, dat_red_err])

df_temp2 = pd.read_csv('./'+results_dir2+'/final_results.txt',names=["sensor_geom","threshold","run_iter","sig_eff","sig_eff_err","bg_rej","bg_rej_err","dat_red","dat_red_err"])
for size_iter in sizes:
    for threshold in thresholds:
        sig_eff_errStat = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["sig_eff_err"].max()
        sig_eff_errML = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["sig_eff"].std()
        sig_eff_err = np.sqrt(sig_eff_errML**2 + sig_eff_errStat**2)
        sig_eff = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["sig_eff"].mean()
        
        bg_rej_errStat = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["bg_rej_err"].max()
        bg_rej_errML = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["bg_rej"].std()
        bg_rej_err = np.sqrt(bg_rej_errML**2 + bg_rej_errStat**2)
        bg_rej = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["bg_rej"].mean()
        
        dat_red_errStat = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["dat_red_err"].max()
        dat_red_errML = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["dat_red"].std()
        dat_red_err = np.sqrt(dat_red_errML**2 + dat_red_errStat**2)
        dat_red = df_temp2[(df_temp2.sensor_geom==size_iter) & (df_temp2.threshold==threshold)]["dat_red"].mean()
        dictionary2[size_iter].append([threshold, sig_eff, sig_eff_err, bg_rej, bg_rej_err, dat_red, dat_red_err])

for (geom, values), (geom2, values2) in zip(dictionary.items(), dictionary2.items()):
    thresh = []
    se = []
    se_err = []
    bgr = []
    bgr_err = []
    dr = []
    dr_err = []
    for thresh_iter in values:
        thresh.append(float(thresh_iter[0]))
        se.append(100*float(thresh_iter[1]))
        se_err.append(100*float(thresh_iter[2]))
        bgr.append(100*float(thresh_iter[3]))
        bgr_err.append(100*float(thresh_iter[4]))
        dr.append(100*float(thresh_iter[5]))
        dr_err.append(100*float(thresh_iter[6]))
        
    thresh2 = []
    se2 = []
    se_err2 = []
    bgr2 = []
    bgr_err2 = []
    dr2 = []
    dr_err2 = []
    for thresh_iter2 in values2:
        thresh2.append(float(thresh_iter2[0]))
        se2.append(100*float(thresh_iter2[1]))
        se_err2.append(100*float(thresh_iter2[2]))
        bgr2.append(100*float(thresh_iter2[3]))
        bgr_err2.append(100*float(thresh_iter2[4]))
        dr2.append(100*float(thresh_iter2[5]))
        dr_err2.append(100*float(thresh_iter2[6]))
        
    geom_tmp = legend_dict[geom]
    plt.errorbar(thresh,se,yerr=se_err, capsize=5 ,fmt='-', color=colors[1])
    plt.scatter(thresh,se, s=15, marker='o', color=colors[1],label="With Lorentz drift")
    plt.errorbar(thresh2,se2,yerr=se_err2, capsize=5 ,fmt='-.', color=colors[0])
    plt.scatter(thresh2,se2, s=15, marker='o', color=colors[0],label="Without Lorentz drift")
    plt.legend(fontsize=12, title_fontsize=12)
    plt.xlabel("Low/high pT boundary [GeV]", fontsize=14)
    plt.ylabel("Signal efficiency [%]", fontsize=14)
    plt.title(f'Signal efficiency for {legend_dict[geom]} um3', fontsize=14)
    plt.grid()
    plt.ylim(70,100)
    # plt.show()
    plt.tight_layout()
    plt.savefig(f'{save_dir}final-signal-efficiency_{geom}.png')
    plt.close()

    plt.errorbar(thresh,bgr,yerr=bgr_err, capsize=5 ,fmt='-', color=colors[1])
    plt.scatter(thresh,bgr, s=15, marker='o', color=colors[1],label="With Lorentz drift")
    plt.errorbar(thresh2,bgr2,yerr=bgr_err2, capsize=5 ,fmt='-.', color=colors[0])
    plt.scatter(thresh2,bgr2, s=15, marker='o', color=colors[0],label="Without Lorentz drift")
    plt.legend(fontsize=12, title_fontsize=12)
    plt.xlabel("Low/high pT boundary [GeV]", fontsize=14)
    plt.ylabel("Background rejection [%]", fontsize=14)
    plt.title(f'Background rejection for {geom} um3', fontsize=14)
    plt.grid()
    plt.ylim(10,60)
    # plt.show()
    plt.tight_layout()
    plt.savefig(f'{save_dir}final-background-rejection_{geom}.png')
    plt.close()

    plt.errorbar(thresh,dr, yerr=dr_err, capsize=5 ,fmt='-', color=colors[1])
    plt.scatter(thresh,dr, s=15, marker='o', color=colors[1],label="With Lorentz drift")
    plt.errorbar(thresh2,dr2, yerr=dr_err2, capsize=5 ,fmt='-.', color=colors[0])
    plt.scatter(thresh2,dr2, s=15, marker='o', color=colors[0],label="Without Lorentz drift")
    plt.legend(fontsize=12, title_fontsize=12)
    plt.xlabel("Low/high pT boundary [GeV]", fontsize=14)
    plt.ylabel("Data reduction [%]", fontsize=14)
    plt.title(f'Data reduction for {geom} um3', fontsize=14)
    plt.grid()
    plt.ylim(10,60)
    # plt.show()
    plt.tight_layout()
    plt.savefig(f'{save_dir}final-data-reduction_{geom}.png')
    plt.close()
print("Run complete.")
```
